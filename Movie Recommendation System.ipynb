##### Importing ast module and using the literal_eval method that will convert the str into list 

```python
import ast
# ast.literal_eval(mov["genres"][0])
```

```python
list_of_unformated_columns = ["genres","keywords","cast"]
```

```python
# function to extract names from dicts and get them in the list for each row

def proper(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i["name"])
    return L
```

```python
# mov["genres"] = mov["genres"].apply(proper)
```

```python
mov["crew"][1]
```

```python
for item in list_of_unformated_columns:
    mov[item] = mov[item].apply(proper)    
     
    
```

#### Extracting director for each movie from crew

```python
mov["crew"] = mov["crew"].apply(lambda x : [i["name"] for i in ast.literal_eval(x) if i["job"]=="Director"])
```

##### We need to perform stemming as there are several words that are repetative i.e. "action and actions" or activity or activities.

```python
pip install nltk
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
ps.stem("loving")
```

```python
ps.stem("loved")
```

```python
def stem(text):
    lst=[]
    for i in text.split():
        lst.append(ps.stem(i))
    return " ".join(lst)
```

```python
mov["tags"] = mov["tags"].apply(stem)
```

```python
mov["tags"][1]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(mov["tags"]).toarray()
```

```python
cv.get_feature_names()
```

### Calculating angular distance (cosine distance) of each vector (each movie) with all the movies in the dataframe

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
similarity[1]
```

##### Function for recommendation 

```python
def recommend(movie):
    movie_index = mov[mov["title"] == movie].index[0]
    distances = similarity[movie_index]
    recommended_movies = sorted(list(enumerate(distances)), reverse = True, key = lambda x:x[1])[1:6]
    
    for i in recommended_movies:
        print(mov.iloc[i[0]].title)
```

```python
recommend("Avatar")
```

```python
recommend("Batman Begins")
```

```python
recommend("Spectre")
```

```python
recommend("Krrish")
```
